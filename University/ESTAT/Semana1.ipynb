Exercício 1

```python
# Valores de entrada

valor_empréstimo = float(input("Valor empréstimo: "))
salario_comprador = float(input("Salário do comprador: "))
anos = int(input("Duração de empréstimo (anos): "))

# Prestação mensal

prestação = salario_comprador / (12 * anos)

# Aprovação do empréstimo

if prestação > salario_comprador * 0.30:
    print("Lamento mas o empréstimo não foi aceite!")
else:
    print("O seu empréstimo foi aprovado!")
    print(f"A prestação mensal será de {prestação:.1f} euros, durante {anos} anos.")
```

Exercício 4

```python
# Valores de entrada

lista = list()

while True:
    num = int(input("Insira um número: "))
    if num < 0:
        break
    else:
        if num not in lista:
            lista.append(num)

if len(lista) == 0:
    print('Lista vazia, não é possível obter mais dados!')
else:
    maior_valor = max(lista)
    indice_maior = lista.index(maior_valor)
    menor_valor = min(lista)
    indice_menor = lista.index(menor_valor)
    nrepetidos = len(lista)
    print(f'O nº de elementos não repetidos é: {nrepetidos}, o maior valor e o seu indíce são respetivamente: {maior_valor} e {indice_maior}, o menor valor e o seu indíce são respetivamente: {menor_valor} e {indice_menor}')
```

Exercício 6

```python
# Valores de entrada

lista = list()
lista_par = list()
lista_impar = list()

# Ciclo do algoritmo

for i in range(10):
    num = int(input("Insira um número: "))
    lista.append(num)
    if num % 2 == 0:
        lista_par.append(num)
    else:
        lista_impar.append(num)

# Sorteamento das listas

lista_par.sort(reverse = True)
lista_impar.sort(reverse = True)

# Output

print(f'Esta é a lista completa: {lista}.')
print(f'Esta é a lista dos números pares: {lista_par}.')
print(f'Esta é a lista dos números ímpares: {lista_impar}.')
```
