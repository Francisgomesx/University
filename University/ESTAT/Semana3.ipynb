```python
# Valores de entrada
prob_maquina_A = 0.45
prob_maquina_B = 0.55
prob_defeito_A = 0.10
prob_defeito_B = 0.20
n = 10

# Cálculo

p = prob_defeito_A * prob_maquina_A + prob_defeito_B * prob_maquina_B
p1 = 1 -stats.binom.cdf(1, n, p)
print(f'A probabilidade de haver duas ou mais peças defeituosas é {p1:.3f}.')

p2 = stats.binom.pmf(0, n, p) + stats.binom.pmf(1, 10, p)
print(f'A probabilidade de haver menos de duas peças defeituosas é {p2:.3f}.')

E = 2.5 * p1 + 5 * p2
print(f'O valor esperado do preço, por peça, é {E:.2f} euros.')
```
